```python
import pandas as pd
import numpy as np
import names
import random
```

```python
def sample_from_dict(d, sample=10):
    keys = random.sample(list(d), sample)
    values = [d[k] for k in keys]
    return dict(zip(keys, values))
```

```python
nobs = 1000
nsam = int(0.5*nobs)
ntotal = 3*nsam
npits = 100
ncat = 10
```

```python
pessoas = [names.get_full_name() for i in range(nobs)]
ids = [i for i in range(nobs)]
users = {ids[i]: pessoas[i] for i in range(len(ids))}
users
```

```python
u1 =sample_from_dict(users, sample=nsam)
u2 =sample_from_dict(users, sample=nsam)
u3 =sample_from_dict(users, sample=nsam)
u1 = list(u1.items())
u2 = list(u2.items())
u3 = list(u3.items())
users_sam = u1+u2+u3
users_sam
```

```python
users_sam_ids = [users_sam[i][0] for i in range(ntotal)]
users_sam_pessoas = [users_sam[i][1] for i in range(ntotal)]
pits = np.random.choice(list(range(1, npits)), ntotal)
categorias = np.random.choice(list(range(1, ncat)), ntotal)
horas = [random.randint(0,24) for i in range(ntotal)]
comprou = [random.randint(0,1) for i in range(ntotal)]
preco = [np.round(random.random()*1000, 2) for i in range(ntotal)]
quantidade = [random.randint(0,4) for i in range(ntotal)]
```

```python
dados = {
    "nome":  users_sam_pessoas,
    "id": users_sam_ids,
    "pit": pits,
    "categoria": categorias,
    "comprou": comprou,
    "preco": preco,
    "quantidade": quantidade,
    "horas": horas
}
```

```python
df = pd.DataFrame(dados)
df.head()
```
